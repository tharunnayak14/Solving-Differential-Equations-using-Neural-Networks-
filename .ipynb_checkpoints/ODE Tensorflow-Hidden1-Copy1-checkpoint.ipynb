```python
import tensorflow as tf
import matplotlib.pyplot as plt
import numpy as np
```

```python
# initial condition
f0 = 0
# infinitesimal small number
inf_s = np.sqrt(np.finfo(np.float32).eps)

# Parameters
learning_rate = 0.001
training_steps = 1000
batch_size = 100
display_step = training_steps/100

# Network Parameters
n_input = 1     # input layer number of neurons
n_hidden_1 = 32 # 1st layer number of neurons
n_output = 1    # output layer number of neurons

weights = {
    'h1': tf.Variable(tf.random.normal([n_input, n_hidden_1])),
    'out': tf.Variable(tf.random.normal([n_hidden_1, n_output]))
}
biases = {
    'b1': tf.Variable(tf.random.normal([n_hidden_1])),
    'out': tf.Variable(tf.random.normal([n_output]))
}

# Stochastic gradient descent optimizer.
optimizer = tf.optimizers.Adam(learning_rate)
```

```python
# Create model
def multilayer_perceptron(x):
    x = np.array([[[x]]],  dtype='float32')
    # Hidden fully connected layer with 32 neurons
    layer_1 = tf.add(tf.matmul(x, weights['h1']), biases['b1'])
    layer_1 = tf.nn.sigmoid(layer_1)
    tf.nn.batch_normalization()
    
    # Output fully connected layer
    output = tf.matmul(layer_1, weights['out']) + biases['out']
    tf.nn.batch_normalization()
    return tf.nn.sigmoid(output)

def g(x):
    return x * multilayer_perceptron(x) + f0

def sin(x):
    return np.sin(x)
    
# Custom loss function to approximate the derivatives
def custom_loss_sin():
    summation = []
    for x in np.linspace(0,10,100):
        dNN = (g(x+inf_s)-g(x))/inf_s
        summation.append((dNN - sin(x))**2)
    return tf.reduce_sum(tf.abs(summation))
    # return tf.sqrt(tf.reduce_mean(tf.abs(summation)))
```

```python
def train_step():
    with tf.GradientTape() as tape:
        loss = custom_loss_sin()
    trainable_variables = list(weights.values()) + list(biases.values())
    gradients = tape.gradient(loss, trainable_variables)
    optimizer.apply_gradients(zip(gradients, trainable_variables))
```

```python
for i in range(training_steps):
    train_step()
    if i % display_step == 0:
        print("loss: %f " % (custom_loss_sin()))
```

```python
from matplotlib.pyplot import figure

figure(figsize=(6, 8))


# True Solution (found analitically)
def true_solution_sin(x):
    return -np.cos(x)+1


X = np.linspace(0, 10, 100)
result = []
for i in X:
    # result.append(f(i))
    print(i)
    result.append(g(i).numpy()[0][0][0])
    print(result)

S = true_solution_sin(X)

plt.plot(X, S, label="Original Function")
plt.plot(X, result, label="Neural Net Approximation")
plt.legend(loc=2, prop={'size': 20})
plt.show()
```

## Tensorflow.GradientTape instead of Manually computing the finite difference 

The same code, but now we are using `tf.GradientTape` to automatically take the derivative for us:

```python
weights = {
    'h1': tf.Variable(tf.random.normal([n_input, n_hidden_1])),
    'out': tf.Variable(tf.random.normal([n_hidden_1, n_output]))
}
biases = {
    'b1': tf.Variable(tf.random.normal([n_hidden_1])),
    'out': tf.Variable(tf.random.normal([n_output]))
}


def train_step_t():
    X = np.linspace(-1, 1, 10)
    with tf.GradientTape(persistent=True) as tape:
        summation = []
        for x in X:
            x = tf.constant([[[x]]],  dtype='float32')
            tape.watch(x)
            g_x = x * multilayer_perceptron(x) + f0
            dNN = tape.gradient(g_x, x)
            summation.append(dNN - sin(x))
        loss = tf.reduce_sum(tf.abs(summation))
    trainable_variables = list(weights.values()) + list(biases.values())
    gradients = tape.gradient(loss, trainable_variables)
    optimizer.apply_gradients(zip(gradients, trainable_variables))


for i in range(training_steps):
    train_step_t()
    if i % display_step == 0:
        print("loss: %f " % (custom_loss_sin()))
```
