```python
import tensorflow as tf
import matplotlib.pyplot as plt
import numpy as np
```

```python
# initial condition
A = 0
# infinitesimal small number
inf_s = np.sqrt(np.finfo(np.float32).eps)

# Parameters
learning_rate = 0.01
training_steps = 100
batch_size = 100
display_step = training_steps/10

# Network Parameters
n_input = 1     # input layer number of neurons
n_hidden_1 = 50 # 1st layer number of neurons
n_hidden_2 = 50 # 2nd layer number of neurons
n_output = 1    # output layer number of neurons

weights = {
    'h1': tf.Variable(tf.random.normal([n_input, n_hidden_1])),
    'h2': tf.Variable(tf.random.normal([n_hidden_1, n_hidden_2])),
    'out': tf.Variable(tf.random.normal([n_hidden_2, n_output]))
}
biases = {
    'b1': tf.Variable(tf.random.normal([n_hidden_1])),
    'b2': tf.Variable(tf.random.normal([n_hidden_2])),
    'out': tf.Variable(tf.random.normal([n_output]))
}

# Stochastic gradient descent optimizer.
optimizer = tf.optimizers.Adam(learning_rate)
```

```python
# Create model
def multilayer_perceptron(x):
    x = np.array([[[x]]],  dtype='float32')
    # Hidden fully connected layer with 32 neurons
    layer_1 = tf.add(tf.matmul(x, weights['h1']), biases['b1'])
    layer_1 = tf.nn.sigmoid(layer_1)
    # Hidden fully connected layer with 32 neurons
    layer_2 = tf.add(tf.matmul(layer_1, weights['h2']), biases['b2'])
    layer_2 = tf.nn.sigmoid(layer_2)
    # Output fully connected layer
    output = tf.matmul(layer_2, weights['out']) + biases['out']
    return tf.nn.sigmoid(output)

# Universal Approximator
def g(x):
    return x * multilayer_perceptron(x) + A

# Given EDO
def f(x):
    return np.exp(-x/5)*np.cos(x)-g(x)/5

# Custom loss function to approximate the derivatives
def custom_loss_1():
    summation = []
    for x in np.linspace(0, 2, 50):
        dNN = (g(x+inf_s)-g(x))/inf_s
        summation.append((dNN - f(x))**2)
    return tf.reduce_sum(tf.abs(summation))
#     X = np.linspace(0, 2, 10)
#     with tf.GradientTape(persistent=True) as tape:
#         summation = []
#         for x in X:
#             x = tf.constant([[[x]]],  dtype='float32')
#             tape.watch(x)
#             g_x = x * multilayer_perceptron(x) + A
#             dNN = tape.gradient(g_x, x)
#             summation.append((dNN - f(x))**2)
#         loss = tf.reduce_sum(tf.abs(summation))
#     return loss, tape
```

```python
def train_step():
    with tf.GradientTape() as tape:
        loss = custom_loss_1()
    trainable_variables = list(weights.values()) + list(biases.values())
    gradients = tape.gradient(loss, trainable_variables)
    optimizer.apply_gradients(zip(gradients, trainable_variables))
```

```python
for i in range(training_steps):
    train_step()
    if i % display_step == 0:
        print("loss: %f " % (custom_loss_1()))
```

```python
from matplotlib.pyplot import figure

figure(figsize=(10, 10))
# True Solution (found analitically)


def true_solution(x):
    return np.exp(-x/5)*np.sin(x)


X = np.linspace(0, 2, 100)
result = []
for i in X:
    result.append(g(i).numpy()[0][0][0])

S = true_solution(X)

plt.plot(X, S, label="Original Function", linestyle='--')
plt.plot(X, result, label="Neural Net Approximation")
plt.legend(loc=2, prop={'size': 20})
plt.show()
```

```python
def g(x):
    return A*(1-x)+B*x+x*(1-x)+x*(1-x)*multilayer_perceptron(x)

# Given EDO


def f(x):
    return -1

# Custom loss function to approximate the derivatives


def custom_loss_2():
    X = np.linspace(0, 2, 10)
    with tf.GradientTape(persistent=True) as tape2:
        summation = []
        for x in X:
            x = tf.constant([[[x]]],  dtype='float32')
            tape2.watch(x)
            with tf.GradientTape() as tape:
                tape.watch(x)
                g_x = x * multilayer_perceptron(x) + A
            dNN = tape.gradient(g_x, x)
        dNN = tape.gradient(dNN, x)
        summation.append((dNN - f(x))**2)
        loss = tf.reduce_sum(tf.abs(summation))
    return loss, tape
```

```python
def train_step():
    loss = custom_loss_2()[0]
    trainable_variables = list(weights.values()) + list(biases.values())
    gradients = tape.gradient(loss, trainable_variables)
    optimizer.apply_gradients(zip(gradients, trainable_variables))
```

```python
for i in range(training_steps):
    train_step()
    if i % display_step == 0:
        print("loss: %f " % (custom_loss_2()[0]))
```
